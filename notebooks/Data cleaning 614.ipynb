```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Step 1: Group rare categories in `43A - Main Span Material`
common_materials = [
    'Concrete', 'Steel', 'Concrete Continuous', 'Steel Continuous', 
    'Prestressed Concrete Continuous', 'Prestressed Concrete'
]
data['43A - Main Span Material'] = data['43A - Main Span Material'].apply(
    lambda x: x if x in common_materials else 'Other'
)

# Step 2: Group rare categories in `43B - Main Span Design`
common_designs = [
    'Tee Beam', 'Stringer/Multi-beam or Girder', 'Culvert', 'Arch - Deck', 
    'Girder and Floorbeam System', 'Box Beam or Girders - Multiple', 'Slab', 'Truss - Thru'
]
data['43B - Main Span Design'] = data['43B - Main Span Design'].apply(
    lambda x: x if x in common_designs else 'Other'
)

# Step 3: One-Hot Encoding for `43A - Main Span Material` and `43B - Main Span Design`
encoder = OneHotEncoder(drop='first', sparse_output=False)

# Fit and transform the Main Span Material
material_encoded = encoder.fit_transform(data[['43A - Main Span Material']])
material_encoded_df = pd.DataFrame(material_encoded, columns=encoder.get_feature_names_out(['43A - Main Span Material']))

# Fit and transform the Main Span Design
design_encoded = encoder.fit_transform(data[['43B - Main Span Design']])
design_encoded_df = pd.DataFrame(design_encoded, columns=encoder.get_feature_names_out(['43B - Main Span Design']))

# Concatenate encoded columns back to the original DataFrame and drop original categorical columns
data = pd.concat([data, material_encoded_df, design_encoded_df], axis=1)
data = data.drop(['43A - Main Span Material', '43B - Main Span Design'], axis=1)

# Display the transformed data
print(data)
```

```python
# Impute missing values for the 'age' column with the mean
data['Age'] = data['Age'].fillna(data['Age'].mean())
```

```python
import numpy as np
import pandas as pd
from scipy import stats
# Columns for further outlier detection
columns_to_check = ['29 - Average Daily Traffic', '49 - Structure Length (ft.)', 'CAT29 - Deck Area (sq. ft.)']

# Calculate z-scores for further filtering
z_scores = np.abs(stats.zscore(data[columns_to_check]))

# Set a higher threshold (e.g., 3.5 or 4) to filter out additional outliers
threshold = 3.5

# Filter rows where any z-score exceeds the threshold
data = data[(z_scores < threshold).all(axis=1)]

# Display the filtered data
print(data)
```

```python
from sklearn.preprocessing import MinMaxScaler

# Define the columns to scale
columns_to_normalize = ['29 - Average Daily Traffic', '49 - Structure Length (ft.)', 'CAT29 - Deck Area (sq. ft.)']

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Apply normalization
data[columns_to_normalize] = scaler.fit_transform(data[columns_to_normalize])

# Display the normalized data
print(data)
```
